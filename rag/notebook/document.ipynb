```python
import os 
os.makedirs("../data/text_files", exist_ok=True)
```

```python
sample_texts = {
    "../data/text_files/python_intro.txt": """
Python Programming Introduction

Python is a high-level, interpreted programming language known for its simplicity
and readability. It supports multiple programming paradigms, including procedural,
object-oriented, and functional programming.

Python was created by Guido van Rossum and was first released in 1991. It has
become one of the most popular programming languages due to its clean syntax and
large ecosystem of libraries.

Python is widely used in web development, data science, artificial intelligence,
machine learning, automation, scripting, cybersecurity, and scientific computing.

Some popular Python frameworks include Django and Flask for web development,
FastAPI for building APIs, and TensorFlow and PyTorch for machine learning.

Example:

def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))

Python's package manager, pip, allows developers to install thousands of
third-party libraries from the Python Package Index (PyPI), making development
faster and more efficient.
""",

    "../data/text_files/machine_learning.txt": """
Machine Learning Introduction

Machine learning is a branch of artificial intelligence (AI) that enables
computers to learn patterns from data without being explicitly programmed for
every task.

Instead of following fixed rules, machine learning models are trained on
datasets to make predictions or decisions. The more high-quality data a model
is trained on, the better it can generalize to unseen examples.

Machine learning is commonly divided into three categories:

- Supervised Learning: Models learn from labeled data. Examples include spam
  detection, house price prediction, and image classification.

- Unsupervised Learning: Models discover hidden patterns in unlabeled data.
  Common tasks include clustering and dimensionality reduction.

- Reinforcement Learning: An agent learns by interacting with an environment
  and receiving rewards or penalties for its actions.

Popular machine learning libraries in Python include scikit-learn, TensorFlow,
PyTorch, and XGBoost.

Typical machine learning workflow:

1. Collect data.
2. Clean and preprocess the data.
3. Select and train a model.
4. Evaluate the model using test data.
5. Deploy the model for real-world use.

Machine learning powers recommendation systems, fraud detection, autonomous
vehicles, medical diagnosis, speech recognition, and natural language
processing applications like ChatGPT.
"""
}
for filepath, content in sample_texts.items():
    with open(filepath, "w", encoding="utf-8") as f:
        f.write(content)
print("Text files created ")
```

```python
### TextLoders 
from langchain_community.document_loaders import TextLoader

loader = TextLoader("../data/text_files/python_intro.txt", encoding="utf-8")
document = loader.load()
document
```

```python
### Directory loader 
from langchain_community.document_loaders import DirectoryLoader

dir_loader = DirectoryLoader(
    "../data/text_files",
    glob="**/*.txt",
    loader_cls= TextLoader,
    loader_kwargs={'encoding':"utf-8"},
    show_progress=False
)
documents = dir_loader.load()
documents
```

```python
### Directory loader 
from langchain_community.document_loaders import DirectoryLoader,PyMuPDFLoader


dir_loader = DirectoryLoader(
    "../data/pdf",
    glob="**/*.pdf",
    loader_cls= PyMuPDFLoader,
    show_progress=False
)
pdf_documents = dir_loader.load()
pdf_documents
```
